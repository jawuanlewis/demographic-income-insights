```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/cleaned_census_data.csv")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Create a boxplot
sns.boxplot(x="income", y="age", data=df, ax=ax1)
ax1.set_title("Age Distribution by Income Level", pad=20)
ax1.set_xlabel("Income")
ax1.set_ylabel("Age")
ax1.grid(True, alpha=0.3)

# Add violin plot for comparison
sns.violinplot(x="income", y="age", data=df, ax=ax2)
ax2.set_title("Age Density by Income Level", pad=20)
ax2.set_xlabel("Income")
ax2.set_ylabel("Age")
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
